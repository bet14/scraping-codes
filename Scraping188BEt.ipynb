```python
import requests
from bs4 import BeautifulSoup

# URL của trang web cần webscrap
url = "https://vnexpress.net/so-hoa/cong-nghe/ai"

# Gửi yêu cầu GET đến trang web và lấy nội dung HTML
response = requests.get(url)
html_content = response.text

# Sử dụng BeautifulSoup để phân tích nội dung HTML
soup = BeautifulSoup(html_content, "html.parser")
# Lấy nội dung HTML đã được định dạng một cách đẹp mắt và chia thành các dòng
formatted_html = soup.prettify()

# Chia thành các dòng
lines = formatted_html.split('\n')

# In ra 10 dòng đầu tiên
for line in lines[:10]:
    print(line)

```
